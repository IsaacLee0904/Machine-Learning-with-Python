## Q1. & Q2. Naive Bayes
### with candy_data
+ 請使用candy-data.csv資料並以樸素貝葉斯 (Naive Bayes) 來做預測，預測winpercent的結果，若大於50則是1若小於等於50則為0。同樣以['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 'nougat', 'crispedricewafer', 'hard', 'bar', 'pluribus', 'sugarpercent', 'pricepercent']作為預測的分析項目。請以test_size=0.25, random_state=1作為條件。

### Step1.Load the data

```python
candy = pd.read_csv(r'/Users/wen/Desktop/應用機器學習於Python/hw_data/hw2&hw3/candy-data.csv')
candy = candy.iloc[: ,1:13] # 將目標變數切割出來
# 模型參數 x 
candyx = candy.iloc[: , 0:11]
candyx # ML2 x
# 模型預測目標 y
candyy = candy.iloc[: , 11]
candyy # ML2 y
```

### Step2. Sandard

```python
# Z-score
from sklearn.preprocessing import StandardScaler
candy_scale = StandardScaler().fit_transform(candyx)
finalcandy = pd.DataFrame(data=candy_scale, columns=['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 'nougat', 'crispedricewafer', 'hard',
                     'bar', 'pluribus', 'sugarpercent', 'pricepercent'])
finalcandy
```

### Step3. Split to training group and testing group

```python
# 切割資料
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(candyx, candyy, # x,y
                                                    test_size=0.25, 
                                                    random_state=1) # 設定seed，固定每次的訓練結果
# 檢查分割後的形狀
print('x_train:', x_train.shape)
print('y_train:', y_train.shape)
print('x_test:', x_test.shape)
print('y_test:', y_test.shape)
```

### Step4. Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

NB = GaussianNB()
NB.fit(x_train.astype('int'), y_train.astype('int'))
pred1 = NB.predict(x_test)
pred1

# 印出預測結果
predNB = pd.DataFrame({'actual_y':y_test,'pred_y':pred1}) 
predNB.sort_index()
```

### Step5. 衡量績效
​
+ 分類模型的衡量指標與迴歸模型不同，因為分類問題的猜測結果有四種情況：
    + 真實為1, 預測為1：True Positive, TP。
    + 真實為0, 預測為1：False Positive, FP，又稱「偽陽性」或統計學上的型一錯誤(Type I Error)。
    + 真實為1, 預測為0：False Negative, FN)，又稱「偽陰性」或統計學上的型二錯誤(Type II Error)。
    + 真實為0, 預測為0：True Negative, TN。
+ 上述四種情況通常會被畫成2x2的矩陣，又稱為「混淆矩陣(Confusion Matrix)」
+ 而分類指標的計算就圍繞著這個矩陣進行不同角度的計算(詳細請參考CH5)
+ 通常商業決策會將混淆矩陣衍生出的指標，搭配兩種猜錯情況所造成的「成本」一併考慮，因此沒有一定要使用哪種指標。

```python
# Recode preML2
def dummy(predNB): # 自行定義一個分類的function
    predNB.loc[(predNB['actual_y'] > 0) & (predNB['actual_y'] <= 50), 'actual_y'] = 0
    predNB.loc[predNB['actual_y'] > 50, 'actual_y'] = 1
    return predNB
dummy(predNB)

def dummy(predNB): # 自行定義一個分類的function
    predNB.loc[(predNB['pred_y'] > 0) & (predNB['pred_y'] <= 50), 'pred_y'] = 0
    predNB.loc[predNB['pred_y'] > 50, 'pred_y'] = 1
    return predNB
dummy(predNB)

predNB.sort_index()
```

```python
# 載入分類模型的衡量指標套件
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc, roc_auc_score, precision_score, recall_score, f1_score
# 印出混淆矩陣
cm = confusion_matrix(predNB['actual_y'], predNB['pred_y'])
pd.DataFrame({'預測為0': cm[:,0], '預測為1': cm[:,1]}, index=['真實為0','真實為1']) 
```

```python
# 印出常見指標的分數
fpr, tpr, thresholds = roc_curve(predNB['actual_y'], predNB['pred_y']) 
pd.DataFrame({'Accuracy': accuracy_score(predNB['actual_y'], predNB['pred_y']), 'AUC': auc(fpr, tpr), 'Precision': precision_score(predNB['actual_y'], predNB['pred_y']), 'Recall': recall_score(predNB['actual_y'], predNB['pred_y']), 'F1': f1_score(predNB['actual_y'], predNB['pred_y']), 'TPR': tpr[1], 'FNR': cm[1][0]/(cm[1][0]+cm[1][1])}, index=['value:'])
```
